### LinkedList

```python
from LinkedList import LinkedList
```

```python
# Test 1: Initialize empty list
lst = LinkedList()
assert len(lst) == 0
assert str(lst) == "[]"

# Test 2: append to empty list
lst.append(1)
assert len(lst) == 1
assert str(lst) == "[1]"

# Test 3: append multiple elements
lst.append(2)
lst.append(3)
assert len(lst) == 3
assert str(lst) == "[1, 2, 3]"

# Test 4: prepend to empty list
lst2 = LinkedList()
lst2.prepend(10)
assert len(lst2) == 1
assert str(lst2) == "[10]"

# Test 5: prepend to non-empty list
lst2.prepend(20)
lst2.prepend(30)
assert len(lst2) == 3
assert str(lst2) == "[30, 20, 10]"

# Test 6: find existing element
assert lst.find(2) is not None
assert lst.find(2).value == 2

# Test 7: find non-existent element
assert lst.find(999) is None

# Test 8: insert after existing element
lst.insert(15, 2)  # insert 15 after 2
assert len(lst) == 4
assert str(lst) == "[1, 2, 15, 3]"

# Test 9: insert after last element (should update tail)
lst.insert(99, 3)
assert len(lst) == 5
assert str(lst) == "[1, 2, 15, 3, 99]"
assert lst.tail.value == 99

# Test 10: insert after non-existent element
result = lst.insert(100, 999)
assert result == -1  # should return -1
assert len(lst) == 5  # length unchanged

# Test 11: delete existing element
lst.delete(15)
assert len(lst) == 4
assert str(lst) == "[1, 2, 3, 99]"

# Test 12: delete first element
lst.delete(1)
assert len(lst) == 3
assert str(lst) == "[2, 3, 99]"

# Test 13: delete last element (should update tail)
lst.delete(99)
assert len(lst) == 2
assert str(lst) == "[2, 3]"
assert lst.tail.value == 3

# Test 14: delete non-existent element
result = lst.delete(999)
assert result == -1  # should return -1
assert len(lst) == 2  # length unchanged

# Test 15: complex test - mix operations
lst3 = LinkedList()
lst3.append(1)
lst3.prepend(0)
lst3.append(2)
lst3.insert(1.5, 1)  # insert 1.5 after 1
assert len(lst3) == 4
assert str(lst3) == "[0, 1, 1.5, 2]"

# Test 16: find after deletion
lst3.delete(1.5)
assert lst3.find(1.5) is None
assert lst3.find(1) is not None

# Test 17: working with empty list
empty_list = LinkedList()
assert len(empty_list) == 0
assert empty_list.find(1) is None
assert empty_list.delete(1) == -1

# Test 18: prepend and append together
lst4 = LinkedList()
lst4.prepend(100)
lst4.append(200)
lst4.prepend(50)
assert str(lst4) == "[50, 100, 200]"
assert lst4.tail.value == 200

# Test 19: multiple insertions after same element
lst5 = LinkedList()
lst5.append(1)
lst5.insert(2, 1)
lst5.insert(3, 2)
assert str(lst5) == "[1, 2, 3]"

# Test 20: delete all elements
lst6 = LinkedList()
lst6.append(1)
lst6.append(2)
lst6.delete(1)
lst6.delete(2)
assert len(lst6) == 0
assert str(lst6) == "[]"
assert lst6.tail is None
```
